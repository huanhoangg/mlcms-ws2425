```python
# here we set the parameters for the Partial Bifurcation Diagram for Logistic Map for r between 0 and 2.

# these parameters are for creating the dynamical system of logistic map
x_start = 0
x_end = 1
x_number_of_points = 1000
xrange = [[x_start, x_end, x_number_of_points]]

discrete = True
r = 0.5

logistic_map = Task41(r, xrange, discrete)

# these parameters will be given to the bifurcation diagram plotting function in the next cell
# note that this plotting function takes a considerable amount of time, you can try reducing r_accuracy or t_end for shorter computing time.
# if you reduce t_end, you should also reduce num_transient_steps according to it
init_state = np.array([0.5])
t_start = 0
t_end = 600
t_eval = np.arange(t_start, t_end)
r_start = 0
r_end = 2
r_accuracy = 0.0001
r_pars = np.arange(r_start, r_end, r_accuracy)
num_transient_steps = 400
critical_points = []
# you uncomment the next line to show critical points
#critical_points = [1]
title = "Partial Bifurcation Diagram for Logistic Map for r between 0 and 2"
save_loc = "visualization/task_4/logistic_map_bifurcation_r_0_2.png"
```

```python
plot_bifurcation_diagram(
    alphas=r_pars,
    system=logistic_map,
    init_state=init_state,
    t_eval=t_eval,
    num_transient_steps=num_transient_steps,
    critical_points=critical_points,
    title=title,
    save_loc=save_loc,
    task_type="numerical"
)
```
