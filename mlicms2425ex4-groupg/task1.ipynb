```python
import matplotlib.pyplot as plt
from utils import create_matrix, compute_eigenvalues, create_vector_field
```

```python
# Different cases for hyperbolic equilibria
matrices = {
    'Stable node': create_matrix(-1, 0, 0, -2),
    'Unstable node': create_matrix(1, 0, 0, 2),
    'Saddle': create_matrix(-1, 0, 0, 1),
}

# Create figure with subplots
plt.figure(figsize=(15, 5))

for i, (name, matrix) in enumerate(matrices.items(), 1):
    plt.subplot(1, 3, i)
    eigs = compute_eigenvalues(matrix)
    X, Y, U, V = create_vector_field(matrix)

    plt.streamplot(X, Y, U, V, density=1.5)
    plt.plot(0, 0, 'k.', markersize=10)
    plt.grid(True)
    plt.xlim(-2, 2)
    plt.ylim(-2, 2)
    plt.title(f'{name}\n, = {eigs[0]:.2f}, {eigs[1]:.2f}')
    plt.axis('equal')

plt.tight_layout()
plt.show()

# Print analysis of topological equivalence
print("\nTopological Equivalence Analysis:")
print("1. The stable and unstable nodes are not topologically equivalent because")
print("   the direction of flow is reversed (one attracts, one repels).")
print("2. Neither node is topologically equivalent to the saddle because")
print("   saddles have different phase space structure with both")
print("   attractive and repulsive directions.")
```
