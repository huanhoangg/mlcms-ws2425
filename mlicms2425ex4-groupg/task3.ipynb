```python
# Andronov-Hopf bifurcation and its three phase diagrams
ode_system = ["alpha * x1 - x2 - x1 * (x1 ** 2 + x2 ** 2)", "x1 + alpha * x2 - x2 * (x1 ** 2 + x2 ** 2)"]
plot_phase_portrait(ode_system, alpha = -1.0)
plot_phase_portrait(ode_system, alpha = 0.0)
plot_phase_portrait(ode_system, alpha = 1.0)
```

```python
# Andronov-Hopf bifurcation and the orbits when alpha = 1.0 with starting points (2,0) and (0.5, 0)
plot_phase_portrait(ode_system, alpha = 1.0, plot_orbit = True, start_pos = (2, 0))
plot_phase_portrait(ode_system, alpha = 1.0, plot_orbit = True, start_pos = (0.5,0))
```
